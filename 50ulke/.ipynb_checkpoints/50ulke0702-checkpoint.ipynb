## 막대그래프를 사용하여 데이터 시각화

```python
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique() #nunique = 특정 열에 있는 고유한 요소의 개수를 나타냄
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 50]]
    columnNames = list(df)
    nGraphRow = int((nCol + nGraphPerRow - 1) / nGraphPerRow) #nGraphPerRow : 표의 각 행마다 그래프의 수를 의미, nCol : 열의 개수를 나타내는 용어
    plt.figure(num = None, figsize = (6 * nGraphPerRow, 8 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i] # iloc : 인덱스 위치 선택
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar() # 이 숫자형이 아닌 경우에만 해당 열의 데이터의 빈도를 막대 그래프로 시각화하는 작업을 수행
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
plotPerColumnDistribution(df1, 10, 5) #plotPerColumnDistribution :각 열에 대한 분포를 시각화하는 작업을 의미
```

## 히트맵을 사용하여 데이터 시각화

```python
def plotCorrelationMatrix(df, graphWidth):
    filename = df.dataframeName
    df = df.dropna('columns') #dropna : 누락된 값 제거
    df = df[[col for col in df if df[col].nunique() > 1]]
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({df.shape[1]}) is less than 2')
        return
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k') # figsize : 그래프나 차트를 그릴 때의 기본적인 도화지
    corrMat = plt.matshow(corr, fignum = 1) # matshow : 행렬 형태의 데이터를 시각적으로 표현하는 함수
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.title(f'Correlation Matrix for {filename}', fontsize=15)
    plt.show()
plotCorrelationMatrix(df1, 8) # 상관계수 행렬 시각화
```

## 산점도 행렬을 사용하여 데이터 시각화

```python
# Scatter and density plots
def plotScatterMatrix(df, plotSize, textSize):
    df = df.select_dtypes(include =[np.number])
    df = df.dropna('columns')
    df = df[[col for col in df if df[col].nunique() > 1]]
    columnNames = list(df)
    if len(columnNames) > 10:
        columnNames = columnNames[:10]
    df = df[columnNames]
    ax = pd.plotting.scatter_matrix(df, alpha=0.75, figsize=[plotSize, plotSize], diagonal='kde') #산점도 행렬 저장
    corrs = df.corr().values  # corr :두 변수 간의 관계의 강도와 방향을 측정하는데 사용되는 값
    for i, j in zip(*plt.np.triu_indices_from(ax, k = 1)): 
        ax[i, j].annotate('Corr. coef = %.3f' % corrs[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=textSize) # annotate : 그래프에 대한 설명을 위한 주석 달기
    plt.suptitle('Scatter and Density Plot')
    plt.show()
plotScatterMatrix(df1, 18, 10)
```
