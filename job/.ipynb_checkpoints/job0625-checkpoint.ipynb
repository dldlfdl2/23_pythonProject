## 막대그래프를 사용하여 데이터 시각화

```python
location_counts = job['Location'].value_counts().head(30)

plt.figure(figsize=(20,10))
location_counts.plot(kind='bar')
plt.title('Top 30 Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
